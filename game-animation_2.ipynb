```python
import numpy as np
from moviepy import ImageClip, CompositeVideoClip, ColorClip
import moviepy.video.fx as fx

def animar_estilo_jogo():
    print("Iniciando Build Segura do Jogo...")
    duracao = 8
    largura_tela, altura_tela = 1920, 1080
    
    # 1. CENÁRIO (Com lógica de proteção contra largura zero)
    img_fundo = ImageClip("foto1.jpg").with_effects([fx.Resize(height=altura_tela)])
    larg_img = img_fundo.size[0]

    def mover_fundo(t):
        # O operador % (módulo) garante que o valor de x 
        # esteja sempre entre 0 e a largura da imagem
        x = int(-150 * t) % larg_img
        # Se a imagem for menor que a tela, movemos apenas em x=0 para evitar erros
        return (int(-x), 'center')

    fundo = (img_fundo.with_duration(duracao)
             .with_position(mover_fundo))

    # 2. PERSONAGEM (Garantindo coordenadas inteiras)
    def flutuar(t):
        # A função seno cria o movimento suave
        oscilacao = 25 * np.sin(2 * np.pi * 0.5 * t)
        # Convertemos para int() para evitar o erro de broadcast
        y_pos = int(450 + oscilacao)
        return ('center', y_pos)

    personagem = (ImageClip("foto2.png")
                  .with_duration(duracao)
                  .with_effects([fx.Resize(width=700)])
                  .with_position(flutuar))

    # 3. INTERFACE (UI)
    barra_fundo = ColorClip(size=(400, 30), color=(40, 40, 40)).with_duration(duracao).with_position((100, 100))
    barra_vida = ColorClip(size=(380, 20), color=(0, 255, 100)).with_duration(duracao).with_position((110, 105))
    
    avatar = (ImageClip("foto2.png")
              .with_duration(duracao)
              .with_effects([fx.Resize(width=90)])
              .with_position((25, 70)))

    # 4. MONTAGEM FINAL
    # DICA: Adicionamos um fundo preto estático por baixo de tudo como segurança
    fundo_preto = ColorClip(size=(largura_tela, altura_tela), color=(0,0,0)).with_duration(duracao)

    video = CompositeVideoClip([
        fundo_preto, 
        fundo, 
        personagem, 
        barra_fundo, 
        barra_vida, 
        avatar
    ], size=(largura_tela, altura_tela))

    print("Renderizando... Esta versão evita o erro de shape (1080,0).")
    video.write_videofile("game_start_fix.mp4", fps=24, codec="libx264")
    print("Level Complete! Verifique o ficheiro: game_start_fix.mp4")

if __name__ == "__main__":
    animar_estilo_jogo()
```
