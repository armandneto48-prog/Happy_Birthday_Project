```python
import numpy as np
import math
from moviepy import ImageClip, CompositeVideoClip, ColorClip, TextClip
import moviepy.video.fx as fx

def animar_estilo_final():
    print("Iniciando a Grande Build: Estilo Donkey Kong & Homenagem...")
    duracao = 8
    L, A = 1920, 1080
    terco = L // 3 # 640px para cada seção

    # --- SEÇÃO 1: ESQUERDA (O JOGO) ---
    # Fundo da foto original
    fundo_esq = (ImageClip("foto1.jpg")
                 .with_duration(duracao)
                 .with_effects([fx.Resize(height=A)])
                 .with_position((0, 0)))
    
    # Personagem flutuando (O seu "GIF" manual)
    def flutuar_esq(t):
        osc = 30 * math.sin(2 * math.pi * 0.5 * t)
        return (100, int(400 + osc))

    personagem = (ImageClip("foto2.png")
                  .with_duration(duracao)
                  .with_effects([fx.Resize(width=400)])
                  .with_position(flutuar_esq))

    # --- SEÇÃO 2: CENTRO (MENSAGEM E ANIMAL) ---
    fundo_centro = (ColorClip(size=(terco, A), color=(0, 0, 0))
                    .with_duration(duracao)
                    .with_position((terco, 0)))

    # Se você tiver um "cachorro.png" ou "gato.png", use aqui. 
    # Vou usar a foto2 pequena saltando como substituto se não tiver o animal.
    def pular_animal(t):
        # Efeito de pulo parabólico
        pulo = -abs(100 * math.sin(math.pi * t)) 
        return (terco + 200, int(700 + pulo))

    animal = (ImageClip("foto2.png") # Substitua por "animal.png" se tiver!
              .with_duration(duracao)
              .with_effects([fx.Resize(width=200)])
              .with_position(pular_animal))

    # NOTA: O texto requer ImageMagick. Se der erro, use uma imagem 'texto.png'
    try:
        txt = (TextClip(text="FELIZ ANIVERSÁRIO,\nPAI!", font_size=50, color='yellow', font='Arial-Bold')
               .with_duration(duracao)
               .with_position((terco + 100, 200)))
    except:
        txt = ColorClip(size=(10,10), color=(0,0,0)).with_duration(duracao) # Fallback silencioso

    # --- SEÇÃO 3: DIREITA (SELVA DONKEY KONG) ---
    fundo_selva = (ColorClip(size=(terco, A), color=(34, 139, 34)) # Verde Floresta
                   .with_duration(duracao)
                   .with_position((terco * 2, 0)))

    # Pista de floresta (Simulada com retângulos de "terra" marrom)
    pista = (ColorClip(size=(terco, 40), color=(139, 69, 19))
             .with_duration(duracao)
             .with_position((terco * 2, 800)))

    avatar_dir = (ImageClip("cat.webp")
                  .with_duration(duracao)
                  .with_effects([fx.Resize(width=250)])
                  .with_position((terco * 2 + 195, 550)))

    # --- MONTAGEM ---
    video = CompositeVideoClip([
        fundo_esq, personagem,      # Coluna 1
        fundo_centro, animal, txt,  # Coluna 2
        fundo_selva, pista, avatar_dir # Coluna 3
    ], size=(L, A))

    print("Renderizando obra-prima...")
    video.write_videofile("homenagem_final_pro.mp4", fps=24, codec="libx264")
    print("Level Max Complete!")

if __name__ == "__main__":
    animar_estilo_final()
```
