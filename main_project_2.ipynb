```python
import numpy as np
import math
from moviepy import ImageClip, CompositeVideoClip, ColorClip
import moviepy.video.fx as fx

def animar_estilo_final_v2():
    print("Iniciando a Build: Ecossistema IT & Gato Patrulheiro...")
    duracao = 8
    L, A = 1920, 1080
    terco = L // 3 

    # --- SEÇÃO 1: ESQUERDA (O JOGO) ---
    fundo_esq = (ImageClip("foto1.jpg")
                 .with_duration(duracao)
                 .with_effects([fx.Resize(height=A)])
                 .with_position((0, 0)))
    
    def flutuar_esq(t):
        osc = 30 * math.sin(2 * math.pi * 0.5 * t)
        return (100, int(400 + osc))

    personagem = (ImageClip("foto2.png")
                  .with_duration(duracao)
                  .with_effects([fx.Resize(width=400)])
                  .with_position(flutuar_esq))

    # --- SEÇÃO 2: CENTRO (ECOSSISTEMA IT CENTRALIZADO) ---
    img_it = ImageClip("it_ecossystem.webp").with_duration(duracao)
    
    # 1. Redimensionamos primeiro pela altura para preencher a tela verticalmente
    img_it = img_it.with_effects([fx.Resize(height=A)])
    
    # 2. Calculamos o corte para pegar o CENTRO da imagem
    largura_atual_da_img = img_it.size[0]
    meio_da_img = largura_atual_da_img // 2
    
    # Queremos um pedaço de largura 'terco' (640px) centrado
    x1 = meio_da_img - (terco // 2)
    x2 = meio_da_img + (terco // 2)
    
    # Aplicamos o Crop (x1, y1, x2, y2)
    fundo_centro = (img_it.with_effects([fx.Crop(x1=x1, y1=0, x2=x2, y2=A)])
                    .with_position((terco, 0)))

    # O resto da seção (animal e texto) permanece igual
    def pular_animal(t):
        pulo = -abs(120 * math.sin(math.pi * 1.5 * t))
        return (terco + 200, int(750 + pulo))

    animal_centro = (ImageClip("cat.webp")
                     .with_duration(duracao)
                     .with_effects([fx.Resize(width=180)])
                     .with_position(pular_animal))

    def pular_animal(t):
        pulo = -abs(120 * math.sin(math.pi * 1.5 * t)) # Pulando um pouco mais rápido
        return (terco + 200, int(750 + pulo))

    animal_centro = (ImageClip("cat.webp")
                     .with_duration(duracao)
                     .with_effects([fx.Resize(width=180)])
                     .with_position(pular_animal))

    # SOLUÇÃO PARA O TEXTO: Usar uma imagem em vez de TextClip
    # Crie um arquivo 'texto.png' com a mensagem para garantir que apareça!
    try:
        mensagem_txt = (ImageClip("texto.png") 
                        .with_duration(duracao)
                        .with_effects([fx.Resize(width=400)])
                        .with_position((terco + 120, 150)))
    except:
        print("Aviso: 'texto.png' não encontrado. Usando fallback vazio.")
        mensagem_txt = ColorClip(size=(1,1), color=(0,0,0)).with_duration(duracao)

    # --- SEÇÃO 3: DIREITA (GATO PATRULHEIRO NA SELVA) ---
    fundo_selva = (ColorClip(size=(terco, A), color=(34, 139, 34))
                   .with_duration(duracao)
                   .with_position((terco * 2, 0)))

    pista = (ColorClip(size=(terco, 40), color=(139, 69, 19))
             .with_duration(duracao)
             .with_position((terco * 2, 850)))

    # LÓGICA DO GATO ANDANDO (Esquerda para Direita em Ciclo)
    def gato_andando(t):
        # O gato percorre a largura do terço (640px) e reseta a cada 4 segundos
        ciclo = 4 
        progresso = (t % ciclo) / ciclo
        x_relativo = progresso * (terco - 150) # -150 para não bater na borda
        return (int(terco * 2 + x_relativo), 650)

    gato_patrulha = (ImageClip("cat.webp")
                     .with_duration(duracao)
                     .with_effects([fx.Resize(width=200)])
                     .with_position(gato_andando))

    # --- MONTAGEM ---
    video = CompositeVideoClip([
        fundo_esq, personagem,           # Coluna 1
        fundo_centro, animal_centro, mensagem_txt, # Coluna 2 (Com imagem it_ecossystem)
        fundo_selva, pista, gato_patrulha # Coluna 3 (Com gato andando)
    ], size=(L, A))

    print("Renderizando versão final com Ecossistema IT...")
    video.write_videofile("homenagem_it_edition.mp4", fps=24, codec="libx264")
    print("Missão Cumprida!")

if __name__ == "__main__":
    animar_estilo_final_v2()
```
