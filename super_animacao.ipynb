```python
from moviepy import ImageClip, CompositeVideoClip
import moviepy.video.fx as fx

def criar_efeito_3d():
    print("Preparando os motores da animação...")
    
    # 1. Carregar as camadas
    # Vamos usar uma duração de 7 segundos para ser mais emocionante
    duracao = 7
    fundo = ImageClip("foto1.jpg").with_duration(duracao)
    pessoas = ImageClip("foto2.png").with_duration(duracao)
    
    largura, altura = fundo.size

    # 2. ANIMAÇÃO DO FUNDO (Sutil)
    # Ele vai crescer de 100% para 110% e deslizar 20 pixels para a esquerda
    fundo_final = (fundo
                   .with_effects([fx.Resize(lambda t: 1 + 0.015 * t)])
                   .with_position(lambda t: (-20 * t/duracao, 'center')))

    # 3. ANIMAÇÃO DAS PESSOAS (Destaque)
    # Elas vão crescer de 100% para 125% e deslizar 30 pixels para a direita
    # Esse movimento oposto (fundo para um lado, pessoas para outro) cria o 3D
    pessoas_final = (pessoas
                     .with_effects([fx.Resize(lambda t: 1 + 0.035 * t)])
                     .with_position(lambda t: (30 * t/duracao, 'center')))

    # 4. MONTAGEM FINAL
    # Vamos centralizar a composição em um "palco" do tamanho da foto
    video = CompositeVideoClip([fundo_final, pessoas_final], size=(largura, altura))

    print("Renderizando a obra de arte... Isso pode demorar um pouquinho.")
    video.write_videofile("homenagem_3D.mp4", fps=24, codec="libx264")
    print("\n--- TUDO PRONTO! Verifique o arquivo: homenagem_3D.mp4 ---")

if __name__ == "__main__":
    criar_efeito_3d()
```
