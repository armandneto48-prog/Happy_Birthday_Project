```python
import os
from rembg import remove
from PIL import Image

def preparar_foto():
    # Nome da foto que você tem na pasta (mude se for diferente)
    input_path = 'foto1.jpg' 
    output_path = 'foto2.png'

    if not os.path.exists(input_path):
        print(f"Erro: Não encontrei o arquivo {input_path} na pasta.")
        print("Certifique-se de que a foto está na pasta Happy_Birthday com o nome correto.")
        return

    print("Iniciando a remoção do fundo... Aguarde um momento.")
    
    try:
        input_image = Image.open(input_path)
        # Remove o fundo usando a IA
        output_image = remove(input_image)
        output_image.save(output_path)
        print(f"Sucesso! Imagem salva como: {output_path}")
    except Exception as e:
        print(f"Ocorreu um erro: {e}")

if __name__ == "__main__":
    preparar_foto()
```
