```python
import numpy as np
from moviepy import ImageClip, CompositeVideoClip, ColorClip
import moviepy.video.fx as fx

def animar_estilo_jogo_v2():
    print("Iniciando Renderização da UI Avançada...")
    duracao = 8
    largura, altura = 1920, 1080 

    # 1. CENÁRIO PRINCIPAL (Lado Esquerdo)
    fundo_jogo = (ImageClip("foto1.jpg")
                  .with_duration(duracao)
                  .with_effects([fx.Resize(height=altura)])
                  .with_position((0, 0)))

    # 2. PAINEL DE INTERFACE (O fundo colorido à direita)
    # Criamos um retângulo que ocupa a lateral direita
    largura_painel = 500
    painel_lateral = (ColorClip(size=(largura_painel, altura), color=(30, 30, 60)) # Azul Escuro
                      .with_duration(duracao)
                      .with_opacity(0.8) # Ligeiramente transparente
                      .with_position((largura - largura_painel, 0)))

    # 3. FUNDO ANIMADO PARA O AVATAR (Acima da foto ambulante)
    # Podemos usar um ColorClip que muda de cor ou uma imagem pequena que roda
    fundo_avatar_animado = (ColorClip(size=(200, 200), color=(255, 50, 50))
                            .with_duration(duracao)
                            .with_position((largura - 350, 150))
                            .with_effects([fx.MirrorX()])) # Exemplo de efeito simples

    # 4. PERSONAGEM (A tua "fotografia ambulante")
    def flutuar(t):
        oscilacao = 20 * np.sin(2 * np.pi * 0.5 * t)
        return (200, 400 + oscilacao) # Posicionado mais à esquerda

    personagem = (ImageClip("foto2.png")
                  .with_duration(duracao)
                  .with_effects([fx.Resize(width=400)])
                  .with_position(flutuar))

    # 5. AVATAR (Fica sobre o fundo animado que criámos)
    avatar = (ImageClip("foto2.png")
              .with_duration(duracao)
              .with_effects([fx.Resize(width=150)])
              .with_position((largura - 325, 175)))

    # 6. MONTAGEM (A ordem aqui define o que fica por cima)
    # O painel_lateral e fundo_avatar entram antes dos elementos de topo
    video = CompositeVideoClip([
        fundo_jogo, 
        painel_lateral, 
        fundo_avatar_animado, 
        personagem, 
        avatar
    ], size=(largura, altura))

    video.write_videofile("game_ui_upgrade.mp4", fps=30, codec="libx264")
    print("Upgrade concluído! Ficheiro: game_ui_upgrade.mp4")

if __name__ == "__main__":
    animar_estilo_jogo_v2()
```

```python
# Substitui as linhas dos assets por estas (ajusta o caminho para a tua pasta real):
caminho_base = r"C:\Users\Armand Neto\Desktop\Happy_Birthday"

cena_esquerda = (VideoFileClip(f"{caminho_base}\\image_0.mp4") 
                 .with_duration(duracao)

import numpy as np
from moviepy import VideoFileClip, CompositeVideoClip, ColorClip
import moviepy.video.fx as fx

def renderizar_jogo_com_video_ui():
    print("Iniciando Renderização com Assets de Vídeo...")
    duracao = 8
    largura_total, altura_total = 1920, 1080 
    largura_jogo_esquerda = 710  
    
    largura_ui = largura_total - largura_jogo_esquerda
    cor_fundo_ui = (0, 100, 200) 

    # --- 1. Assets de VÍDEO (.mp4) ---
    
    # IMPORTANTE: Substituímos ImageClip por VideoFileClip
    # Se o teu vídeo original for curto, usamos .loop() para ele não parar
    cena_esquerda = (VideoFileClip("image_0.mp4") 
                     .with_duration(duracao)
                     .with_effects([fx.Resize(height=altura_total)]) # Garante que preenche a altura
                     .with_effects([fx.Crop(x1=0, y1=0, width=largura_jogo_esquerda, height=altura_total)])
                     .with_position((0, 0)))

    # O teu "Sprite" (se for um vídeo com fundo preto, ele vai aparecer com o fundo)
    sprite_ambulante = (VideoFileClip("foto2_recortada.mp4") 
                        .with_duration(duracao)
                        .with_effects([fx.Resize(height=600)]))

    # --- 2. Painel de Fundo da UI ---
    fundo_painel_ui = (ColorClip(size=(largura_ui, altura_total), color=cor_fundo_ui)
                       .with_duration(duracao)
                       .with_position((largura_jogo_esquerda, 0)))

    # --- 3. Lógica de Movimentação ---
    def flutuar_no_painel(t):
        pos_x = largura_jogo_esquerda + (largura_ui / 4) + (50 * t) 
        oscilacao_y = 30 * np.sin(2 * np.pi * 0.5 * t)
        pos_y = (altura_total / 2) - 100 + oscilacao_y
        return (pos_x, pos_y)

    sprite_animado = sprite_ambulante.with_position(flutuar_no_painel)

    # --- 4. Montagem ---
    video = CompositeVideoClip([
        cena_esquerda,      
        fundo_painel_ui,    
        sprite_animado      
    ], size=(largura_total, altura_total))

    print("Gerando Build final (isto pode demorar um pouco mais com vídeos)...")
    video.write_videofile("jogo_final_com_video.mp4", fps=30, codec="libx264")
    print("Sucesso! Verifica: jogo_final_com_video.mp4")

if __name__ == "__main__":
    renderizar_jogo_com_video_ui()

sprite_ambulante = (VideoFileClip(f"{caminho_base}\\foto2_recortada.mp4") 
                    .with_duration(duracao)
```
