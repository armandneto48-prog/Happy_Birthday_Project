```python
import numpy as np
import math
from moviepy import ImageClip, CompositeVideoClip, ColorClip
import moviepy.video.fx as fx

def animar_estilo_jogo_infinito():
    print("--- GERANDO CENÁRIO INFINITO ---")
    duracao = 8
    largura_tela, altura_tela = 1920, 1080
    velocidade = 150 # pixels por segundo

    # 1. CARREGAR O CENÁRIO
    img_fundo = ImageClip("foto1.jpg").with_effects([fx.Resize(height=altura_tela)])
    largura_foto = img_fundo.size[0]

    # 2. FUNÇÕES DE POSIÇÃO (Garantindo números inteiros e sem erros de cálculo)
    # A primeira imagem
    def pos_fundo1(t):
        x = - (velocidade * t) % largura_foto
        return (int(x), 'center')

    # A segunda imagem (colada na primeira)
    def pos_fundo2(t):
        x = (- (velocidade * t) % largura_foto) + largura_foto
        return (int(x), 'center')

    # 3. APLICAR POSIÇÕES
    fundo_1 = img_fundo.with_duration(duracao).with_position(pos_fundo1)
    fundo_2 = img_fundo.with_duration(duracao).with_position(pos_fundo2)

    # 4. PERSONAGEM (Flutuando)
    def flutuar(t):
        # Usando math.sin para evitar dependência externa se necessário
        oscilacao = 30 * math.sin(math.pi * t)
        return ('center', int(450 + oscilacao))

    personagem = (ImageClip("foto2.png")
                  .with_duration(duracao)
                  .with_effects([fx.Resize(width=700)])
                  .with_position(flutuar))

    # 5. INTERFACE (UI)
    barra_fundo = ColorClip(size=(400, 30), color=(40, 40, 40)).with_duration(duracao).with_position((100, 100))
    barra_vida = ColorClip(size=(380, 20), color=(0, 255, 100)).with_duration(duracao).with_position((110, 105))

    # 6. MONTAGEM FINAL
    video = CompositeVideoClip([
        fundo_1, 
        fundo_2, 
        personagem, 
        barra_fundo, 
        barra_vida
    ], size=(largura_tela, altura_tela))

    print("Renderizando... O erro de ValueError deve estar resolvido!")
    video.write_videofile("game_infinito_v2.mp4", fps=24, codec="libx264")
    print("\n--- SUCESSO! ---")

if __name__ == "__main__":
    animar_estilo_jogo_infinito()
```
