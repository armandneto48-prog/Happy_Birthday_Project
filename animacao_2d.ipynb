```python
from moviepy import ImageClip, TextClip, CompositeVideoClip
import moviepy.video.fx as fx

def criar_estilo_2d():
    print("Criando animação estilo 2D Dinâmico...")
    duracao = 6
    
    # 1. FUNDO: Movimento lateral simples (Pan)
    fundo = (ImageClip("foto1.jpg")
             .with_duration(duracao)
             .with_position(lambda t: (-30 * t, 'center'))) # Desliza 30px por segundo

    # 2. PESSOAS: Entrada com "Slide" e Transparência
    # Elas começam fora da tela (esquerda) e param no centro em 1.5 segundos
    pessoas = (ImageClip("foto2.png")
               .with_duration(duracao)
               .with_position(lambda t: (min(0, -100 + 100*t), 'center')) 
               .with_effects([fx.FadeIn(1.5)])) # Surge suavemente em 1.5s

    # 3. TEXTO (Opcional - Requer ImageMagick instalado no Windows)
    # Se você não tiver o ImageMagick, o código abaixo pode dar erro. 
    # Se der erro, comente as linhas do 'texto' e do 'CompositeVideoClip'
    try:
        texto = (TextClip(text="Feliz Aniversário, Pai!", 
                          font_size=70, color='white', font='Arial-Bold')
                 .with_duration(duracao - 2)
                 .with_start(2) # O texto aparece só depois de 2 segundos
                 .with_position(('center', 0.8), relative=True) # Parte inferior
                 .with_effects([fx.FadeIn(1)]))
        
        elementos = [fundo, pessoas, texto]
    except:
        print("Aviso: TextClip requer ImageMagick. Gerando vídeo apenas com as fotos.")
        elementos = [fundo, pessoas]

    # 4. MONTAGEM
    largura, altura = ImageClip("foto1.jpg").size
    video = CompositeVideoClip(elementos, size=(largura, altura))

    print("Renderizando versão 2D...")
    video.write_videofile("homenagem_2D.mp4", fps=24, codec="libx264")
    print("Sucesso! Veja o arquivo: homenagem_2D.mp4")

if __name__ == "__main__":
    criar_estilo_2d()
```
