```python
import os
from moviepy import ImageClip, CompositeVideoClip
import moviepy.video.fx as fx  # Importante para o resize

def criar_animacao():
    # 1. Carregar as imagens
    fundo = ImageClip("foto1.jpg").with_duration(5)
    pessoas = ImageClip("foto2.png").with_duration(5)

    largura, altura = fundo.size

    # 2. Criar efeito de Zoom no Fundo
    # Na v2.0, usamos .with_effects() com fx.Resize
    fundo_animado = (fundo
                     .with_effects([fx.Resize(lambda t: 1 + 0.02 * t)])
                     .with_position('center'))

    # 3. Criar efeito de Zoom nas Pessoas
    pessoas_animadas = (pessoas
                        .with_effects([fx.Resize(lambda t: 1 + 0.04 * t)])
                        .with_position('center'))

    # 4. Juntar tudo
    video_final = CompositeVideoClip([fundo_animado, pessoas_animadas], size=(largura, altura))

    # 5. Salvar o vídeo
    print("Renderizando o vídeo de homenagem...")
    video_final.write_videofile("homenagem_pai.mp4", fps=24, codec="libx264")

if __name__ == "__main__":
    criar_animacao()
```
